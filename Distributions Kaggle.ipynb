```python
# https://www.kaggle.com/code/hamelg/python-for-data-22-probability-distributions/notebook

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd

```

Normal Distribution

```python
#norm = normal distribution
#mean = 0
#std = 1
#cdf = probability to the left of x
prob_under_minus1 = stats.norm.cdf(x= -1,  
                                loc = 0,               
                                scale= 1)     

prob_over_1 = 1 - stats.norm.cdf(x= 1,  
                                loc = 0,               
                                scale= 1) 

between_prob = 1-(prob_under_minus1+prob_over_1)

print(prob_under_minus1, prob_over_1, between_prob)
```

```python
# Plot normal distribution areas*
#standard deviation = 1 here
#mean = 0

plt.rcParams["figure.figsize"] = (9,9)
                                  
plt.fill_between(x=np.arange(-6,-1,.01), 
                 y1= stats.norm.pdf(np.arange(-6,-1,.01)) ,
                 facecolor='red',
                 alpha=0.35)

plt.fill_between(x=np.arange(1,6,0.01), 
                 y1= stats.norm.pdf(np.arange(1,6,0.01)) ,
                 facecolor='red',
                 alpha=0.35)

plt.fill_between(x=np.arange(-1,1,0.01), 
                 y1= stats.norm.pdf(np.arange(-1,1,0.01)) ,
                 facecolor='blue',
                 alpha=0.35)

plt.text(x=-1.8, y=0.03, s= round(prob_under_minus1,3))
plt.text(x=-0.2, y=0.1, s= round(between_prob,3))
plt.text(x=1.4, y=0.03, s= round(prob_over_1,3));
```

```python
#ppf = std quantile for the 2.5% and 97.5% cutoff
print(stats.norm.ppf(q=0.025))
print(stats.norm.ppf(q=0.975))
```

```python
#cdf = probability to the left of x
prob_25quantile = stats.norm.cdf(x=-1.95996, loc=0, scale=1)
print(prob_25quantile)
```

```python
# Plot normal distribution areas*
#standard deviation = 1 here
#mean = 0

plt.rcParams["figure.figsize"] = (9,9)

# plot 1
# pd.DataFrame(stats.norm.rvs(0,1,100000)).plot(kind="density")

# plot 2
#generate numbers between -4 and 4 with 2000 samples
# plot x and normal distribution of mean=0 and std=1
x=np.linspace(-4,4,2000)
plt.plot(x,stats.norm(0,1).pdf(x))
```

Binomial Distribution
Probability distribution based on outcomes of a given number of random trials 
2 parameters: probability of successes in any given trial and number trials

```python
#samples per trial = 10
#probability of success = 50%
#number of trials = 1000

coin_flips = stats.binom.rvs(n=10, p=.5, size=1000)

#plot 1
pd.DataFrame(coin_flips).hist(range=(-.5,10.5), bins=11)

```

```python
#change the distribution where 80% chance of success
coin_flips = stats.binom.rvs(n=10,p=.8,size=1000)
pd.DataFrame(coin_flips).hist(range=(-0.5,10.5))
```
